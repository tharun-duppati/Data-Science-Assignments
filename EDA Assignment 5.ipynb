```python
import matplotlib.pyplot as plt
import seaborn as sns
# Automatically detect numerical columns
numeric_cols = df.select_dtypes(include='number').columns

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Plot box plots for each numerical column
sns.boxplot(data=df[numeric_cols], orient='h', ax=ax)

# Customize the plot (optional)
plt.title('Box Plots for Numerical Columns')
plt.xlabel('Value')
plt.ylabel('Column Name')
plt.xticks(rotation=45)
plt.show()
```

```python
df["LB"].plot(kind="box",vert=False)
q1=df["LB"].quantile(0.25)
q3=df["LB"].quantile(0.75)
iqr=q3-q1
lw=q1-1.5 * iqr
uw=q3+1.5 * iqr
print("Lower Whisker length:",lw)
print("Upper Whisker length:",uw)
outliers=df[(df["LB"]<lw) | (df["LB"]>uw)]
print("Outliers:",len(outliers))
```
