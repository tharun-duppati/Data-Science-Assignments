Scaling: 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Selecting numerical features
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# Standard Scaling
scaler_standard = StandardScaler()
df[numerical_features] = scaler_standard.fit_transform(df[numerical_features])

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df[numerical_features] = scaler_minmax.fit_transform(df[numerical_features])

print("Scaling applied.")
df.head()
```

```python
'''
1. Standard Scaling :
     -Preferred when the data follows a normal distribution or when the algorithm assumes a Gaussian distribution of features.
     -Useful when features have different units and ranges, but their importance in the model is proportional to their standard deviation.

2. Min-Max Scaling :
     -Preferred when features are on different scales but do not follow a normal distribution.
     -Best for algorithms that are sensitive to the range of data, especially when outliers are not an issue.
'''
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-Hot Encoding for variables with <5 categories which are race and sex
ohe=OneHotEncoder()
df["race"]=ohe.fit_transform(df[["race"]]).toarray()
df["sex"]=ohe.fit_transform(df[["sex"]]).toarray()
```

```python
# Label Encoding for variables with >=5 categories
from sklearn.preprocessing import LabelEncoder
#categorical_label = df_cleaned[['workclass', 'education', 'occupation', 'native_country']]

le=LabelEncoder()
df["workclass"]=le.fit_transform(df["workclass"])
df["education"]=le.fit_transform(df["education"])
df["marital_status"]=le.fit_transform(df["marital_status"])
df["occupation"]=le.fit_transform(df["occupation"])
df["relationship"]=le.fit_transform(df["relationship"])
df["native_country"]=le.fit_transform(df["native_country"])
df["income"]=le.fit_transform(df["income"])
df.head()
```

```python
'''
One-Hot Encoding

Pros:
   -No Assumed Order: It doesn't assume any order between categories, making it great for categorical features like colors or countries.
   -Prevents Misleading Results: It avoids algorithms mistakenly thinking one category is "greater" or "lesser" than another.
Cons:
   -Increases Data Size: It creates extra columns for each category, which can make your dataset larger and harder to work with, especially for features with many categories.
   -Can be Inefficient: It can make models slower and require more memory because of all the new columns created.

Label Encoding

Pros:
   -Efficient: It uses a single column to represent categories with numbers, which saves space and memory.
   -Good for Ordinal Data: It works well when the categories have a natural order, like low, medium, and high.
Cons:
   -Assumes Order: It may incorrectly treat unordered categories as if they have a ranking, which can lead to misleading results.
   -Not Ideal for Some Models: Some algorithms might interpret the numerical values incorrectly (e.g., thinking 1 < 2 < 3 when thereâ€™s no real order).
'''
```
