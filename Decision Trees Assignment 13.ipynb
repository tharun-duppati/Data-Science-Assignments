```python
X_cat=df_final[["sex","cp","fbs","restecg","slope","thal","num"]]
X_cat
```

```python
#------------------------------------------------Feature Engineering---------------------------------------------------
```

```python
#Encoding categorical values
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder,StandardScaler
le=LabelEncoder()
X_cat["sex"]=le.fit_transform(X_cat["sex"])
X_cat["cp"]=le.fit_transform(X_cat["cp"])
X_cat["fbs"]=le.fit_transform(X_cat["fbs"])
X_cat["restecg"]=le.fit_transform(X_cat["restecg"])
X_cat["slope"]=le.fit_transform(X_cat["slope"])
X_cat["thal"]=le.fit_transform(X_cat["thal"])
X_cat["num"]=le.fit_transform(X_cat["num"])
```

```python
#Standardizing ontinuous data
ss=StandardScaler()
X_ss=ss.fit_transform(X_cont)
X_ss=pd.DataFrame(X_ss)
X_ss.columns=list(X_cont)
X_ss
```

```python
#standardized data with label encoding columns

df_final=pd.concat([X_ss,X_cat],axis=1)
df_final
```

```python
#-------------------------------------DecisionTreeClassification---------------------------------------------------------
```

```python
Y=df_final["num"]
X=df_final.drop("num",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(criterion="gini")

model.fit(X_train, Y_train)

# Make predictions on the training set
Y_train_pred = model.predict(X_train)

# Make predictions on the testing set
Y_test_pred = model.predict(X_test)

# Calculate the training accuracy
train_accuracy = accuracy_score(Y_train, Y_train_pred)

# Calculate the testing accuracy
test_accuracy = accuracy_score(Y_test, Y_test_pred)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
# cross validation

training_acc = []
test_acc = []

for i in range(1,101):
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3, random_state=i)
    model = DecisionTreeClassifier(criterion="gini",max_depth=6)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    training_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))

print("Cross validation traning accuracy:", np.round(np.mean(training_acc),2))
print("Cross validation test accuracy:", np.round(np.mean(test_acc),2))
```

```python
#---------------------------------------------------Hyperparameter Tuning---------------------------------------------------
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [1,2,3, 5, 7, 9],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
grid_search.fit(X_train, Y_train)

best_clf = grid_search.best_estimator_
best_clf.fit(X_train, Y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 2))
plot_tree(model, filled=True, feature_names=X.columns)
plt.show()

# prompt: print the number of nodes and depth of the model

# Get the number of nodes and depth of the decision tree
n_nodes = model.tree_.node_count
depth = model.tree_.max_depth

print("Number of nodes:", n_nodes)
print("Depth of the tree:", depth)
```
